```python
x, y, z = img1.shape
print("x=", x, " y=",y)
```

```python
div = 10
xit = x//div
yit = y//div
print("xit=", xit, " yit=",yit)
```

```python
dangerMap = np.zeros((div,div),dtype = np.bool)
print(dangerMap.shape)
print(dangerMap)
```

```python
for X in range(0, div):
    for Y in range(0, div):
        histS1 = cv2.calcHist([img1[X*div:(X+1)*div,Y*div:(Y+1)*div]],[0],None,[256],[0,256])
        histS2 = cv2.calcHist([img2[X*div:(X+1)*div,Y*div:(Y+1)*div]],[0],None,[256],[0,256])
        #print("Histograms img1=", histS1, " img2=", histS2)
        
#         sc = cv2.compareHist(histS1, histS2, cv2.HISTCMP_INTERSECT)
        sc = cv2.compareHist(histS1, histS2, cv2.HISTCMP_BHATTACHARYYA)
        print("Compared Hist at X=", X, " Y=", Y, " : ", sc)
        dangerMap[X,Y] = sc > 0.90
```

```python
print(dangerMap)
print(len(dangerMap))
```
